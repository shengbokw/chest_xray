```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
from scipy.stats import norm
from sklearn.preprocessing import StandardScaler
from scipy import stats
from scipy.stats import norm, skew 
from scipy.special import boxcox1p
```

```python
%matplotlib inline
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'
mpl.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus']=False
```

Load pneumonia_train data

```python
pne_train = pd.read_csv('pneumonia_train_small.csv')
```

Add label for this dataframe

```python
pne_train['label'] = 1
```

Do the same step for normal dataFrame

```python
nor_train = pd.read_csv('normal_train_small.csv')
nor_train['label'] = 0
```

Concatenating the Pneumonia and Normal DataFrame into the whole train dataset

```python
train = pd.concat([pne_train, nor_train], ignore_index=True) 
```

Looking at the Summary for this train data set

```python
train.info()
```

```python
train.head()
```

```python
train.describe()
```

```python
corrmat = train.corr()
f, ax = plt.subplots(figsize=(12, 12))
sns.heatmap(corrmat, vmax=1, square=True);
```

From the Correlation coefficient matrix, We see that for lable, the most positive correlated features are lung_density, gray_14, the most negitive correlated features are gray_0, gray_15, l_lungsize and fraction.

```python
k = 7
cols = corrmat.nlargest(k, 'label')['label'].index
cm = np.corrcoef(train[cols].values.T)
sns.set(font_scale=1.25)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, 
yticklabels=cols.values, xticklabels=cols.values)
plt.show();
```

```python
sns.set()
cols = ['label','gray_14','lung_density','total_gray','gray_6','gray_7','gray_5']
sns.pairplot(train[cols], size = 2.5)
plt.show();
```
